```python
from bs4 import BeautifulSoup
```

```python
def get_headers(input):
    with open(input, "r") as f:
        soup = BeautifulSoup(f, "lxml")
    all_pb = soup.find_all("pb")
    all_headers = []
    for pb in all_pb:
        attr = pb.attrs
        for k in attr:
            if k == "pageheader":
                all_headers.append(attr[k])
    return all_headers
```

```python
# /home/pret-miriad/Documents/JUPYTER/for_headers
file = "/home/pret-miriad/Documents/JUPYTER/for_headers/LOD2M_{}.xml".format("s1_t5")
print(get_headers(file))
```
